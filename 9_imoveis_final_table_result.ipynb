```python
import pandas as pd
import io

texto = """
 | Model                   | R2| RMSE| MAE| MedAE| MAPE |
 |Linear log, inter (all)|	0.63 | 402047.49  | 210913.13	| 127497.35 | 0.22 |
 | RF Numeric Fields (all) |	0.98 | 95300.42	| 57290.74 |	29912.10 |0.07 |
 | RF Num + Yolo (all)| 0.98	| 95801.03	| 57979.46	| 30782.84	| 0.07 |
 | RF Num + text (all)| 0.98 | 93045.62	| 55591.42 | 29443.23	| 0.06 |
 | RF Num+text+yolo (all)| 0.98 |	92842.89 |	55535.63 |	29494.47 | 0.06 |
"""

df = pd.read_csv(io.StringIO(texto), sep='|')
df = df.reset_index(drop=True)
df.drop([' ', 'Unnamed: 7'], axis=1, inplace=True)
df.columns = [x.strip() for x in df.columns]
display(df)
```

```python
df_ML_models = df[ df.Model != 'Linear log, inter (all)' ]

# Calculate the mean of the numerical columns
means = df_ML_models.select_dtypes(include=['number']).mean()

# Create a new row with the calculated means and 'RF means' as the model
new_row = pd.DataFrame({'Model': ['RF models means'], **means})

# Append the new row to the dataframe
df_ML_models = pd.concat([df_ML_models, new_row], ignore_index=True)

display(df_ML_models)
```

```python
df_ML_models_pct = df_ML_models.copy()

for i in range(len(df_ML_models_pct)):
    for j in range(1,len(df_ML_models_pct.columns)):
      df_ML_models_pct.iloc[i, j] = (( df_ML_models_pct.iloc[i,j] - df_ML_models.iloc[-1, j] ) / df_ML_models.iloc[-1, j] * 100).round(2)

df_ML_models_pct.columns = [x + ' %' for x in df.columns]

df_ML_models_pct
```

```python
from IPython.display import display, Latex, HTML
import warnings

def gera_df_stats_latex(df_stats, titulo):

  warnings.filterwarnings('ignore')
  # latex_table = df_stats.style.to_latex(hrules=True)
  latex_table = df_stats.to_latex(index=None)
  # display(Latex(latex_table))

  latex_table = latex_table.replace('tabular','array')
  latex_table = latex_table.replace('toprule','hline')
  latex_table = latex_table.replace('midrule','hline')
  latex_table = latex_table.replace('bottomrule','hline')

  latex_table = '\\textbf{' + titulo + '} \\\\' + latex_table

  warnings.filterwarnings('default')
  return latex_table
```

```python
latex_table = gera_df_stats_latex(df_ML_models_pct,'Results')
display(Latex(latex_table))
```

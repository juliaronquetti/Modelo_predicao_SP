```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Dataframe_25_04.csv', encoding = 'utf-8', sep=';')
df.rename(columns={'index': 'img_index'}, inplace=True)
df.head()

```

```python
df.columns
```

```python
import re

new_columns = []
for column in df.columns:
  new_column = column.lower()
  new_column = re.sub(r'[àáâãäå]', 'a', new_column)
  new_column = re.sub(r'[èéêë]', 'e', new_column)
  new_column = re.sub(r'[ìíîï]', 'i', new_column)
  new_column = re.sub(r'[òóôõö]', 'o', new_column)
  new_column = re.sub(r'[ùúûü]', 'u', new_column)
  new_column = re.sub(r'[ç]', 'c', new_column)
  new_column = new_column.replace(" ", "_")
  new_columns.append(new_column)

df.columns = new_columns
```

```python
cols = list(df.columns)
cols.remove('img_index')
cols.append('img_index')
df = df[cols]
```

```python
df.to_csv('imoveis_raw_url.csv',index=None)
```

```python
df = df.drop(['url', 'imagem_inicial', 'images'], axis=1)
df.to_csv('imoveis_raw.csv',index=None)
```
